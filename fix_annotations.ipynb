```python
def fix_annotation(input_ann_file_path, input_thread_file_path, output_ann_file_path):
    with open(input_ann_file_path, 'r', encoding='utf8') as input_ann_file:
        input_ann_file_content = input_ann_file.read()
        with open(input_thread_file_path, 'r') as input_thread_file:
            input_thread_file_content = input_thread_file.read()
            new_content = []
            _start = 0
            _end = 1
            _in_number = False

            for i, ann_char  in enumerate(input_ann_file_content):
                if not _in_number:
                    if ann_char.isdigit() and (input_ann_file_content[i - 1] == '|' or input_ann_file_content[i - 1] == '.'):
                        _start = i
                        _in_number = True
                    else:
                        new_content.append(ann_char)
                else:
                    if not ann_char.isdigit():
                        _end = i
                        _in_number = False
                        if not input_ann_file_content[_start - 1].isalpha():
                            num = int(input_ann_file_content[_start:_end])
                            # len_content_8 = len(input_thread_file_content[:num])
                            # len_content_16 = len(input_thread_file_content[:num].encode('utf-16le'))
                            # new_num = num - ((len_content_16 - (len_content_8 * 2)) // 2)
                            new_num = len(input_thread_file_content.encode('utf-16le')[:(num * 2)].decode('utf-16le'))

                            new_content.append(str(new_num))
                            new_content.append(ann_char)
                        _start = _end + 1
                        continue

            with open(output_ann_file_path, 'w', encoding='utf8') as fixed_ann:
                fixed_ann.write(''.join(new_content))
```

```python
files = os.listdir(INPUT_ANN_FOLDER)
for file_name in files:
    try:
        fix_annotation(
            INPUT_ANN_FOLDER + '/' + file_name,
            INPUT_THREAD_FOLDER + '/' + file_name,
            OUTPUT_ANN_FOLDER + '/' + file_name
        )
    except:
        print('error')
```
