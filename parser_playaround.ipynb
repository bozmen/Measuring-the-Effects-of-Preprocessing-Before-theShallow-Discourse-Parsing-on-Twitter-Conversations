```python
import json
import pandas as pd
```

```python
results = []

with open('../LinDocker/text/output.json') as f:
    parse_result = json.loads(f.read())
    results = parse_result
```

```python
def to_plain_result(result):
    arg1 = result['Arg1']['RawText']
    arg2 = result['Arg2']['RawText']
    connective = result['Connective']['RawText']
    sense = result['Sense']
    typ = result['Type']
    result_plain = {
        'arg1': arg1,
        'arg2': arg2,
        'connective': connective,
        'sense': sense,
        'type': typ
    }
    return result_plain

def print_args(df, _id): 
    print('arg1:', df.loc[_id].arg1)
    print('arg2:', df.loc[_id].arg2)
    print('connective:', df.loc[_id].connective)
```

```python
plain_results = [to_plain_result(result) for result in results]
```

```python
data = pd.DataFrame.from_dict(plain_results)
```
